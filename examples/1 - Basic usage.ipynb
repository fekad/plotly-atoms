# Usage

```python
from plotly_atoms import Figure
```

## Example 1: Simple call of the function with positions xyz and labels



```python
# The geometry of ethylene C2H4 in Angstrm
xyz = [
    [0.000, 0.000, 0.000],  # C
    [1.340, 0.000, 0.000],  # C
    [-0.545, 0.944, 0.000],  # H
    [-0.545, -0.944, 0.000],  # H
    [1.885, 0.944, 0.000],  # H
    [1.885, -0.944, 0.000],  # H
]

labels = ["C", "C", "H", "H", "H", "H"]
fig = Figure()
fig.add_atoms(xyz, labels)
fig.show()
```

## Example 2: Comparison of styles



```python
# The geometry of ethylene C2H4 in Angstrm
xyz = [
    [0.000, 0.000, 0.000],  # C
    [1.340, 0.000, 0.000],  # C
    [-0.545, 0.944, 0.000],  # H
    [-0.545, -0.944, 0.000],  # H
    [1.885, 0.944, 0.000],  # H
    [1.885, -0.944, 0.000],  # H
]

labels = ["C", "C", "H", "H", "H", "H"]
```

## Example 3: Simple call of the function for water

```python
# The geometry of water H2O in Angstrm
xyz = [
    [0.000, 0.000, 0.000],  # O
    [0.762, -0.584, 0.000],  # H
    [-0.762, -0.584, 0.000],  # H
]

# The labels\elements of the xyz columns
labels = ["O", "H", "H"]

fig = Figure()
fig.add_atoms(xyz,labels)
fig.show()
```

## Example 4: Simple call of the function for iodo acetylene

```python
# The geometry of iodo acetylene ICCH in Angstrm
xyz = [
    [-1.590, 0.000, 0.000],  # I
    [0.000, 0.000, 0.000],  # C
    [1.200, 0.000, 0.000],  # C
    [2.260, 0.000, 0.000],  # H
]
# The labels\elements of the xyz columns
labels = ["I", "C", "C", "H"]
fig = Figure()
fig.add_atoms(xyz, labels)
fig.show()
```
